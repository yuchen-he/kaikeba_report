# This is a program for changing the coloe of a ID photo

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_blue = cv2.imread('./id_photo/blue.jpg', 1)
img_red = cv2.imread('./id_photo/red.jpg', 1)
img_white = cv2.imread('./id_photo/white.jpg', 1)
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(img_blue,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def backgroung_color_blue2white(img):
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # Step1: find a region that the value exceed a threshold, we consider a the blue background
    h = img_hsv[:,:,0]
    s = img_hsv[:,:,1]
    v = img_hsv[:,:,2]
#     img[h>99, h<125, s<256, s>42] = [h_white,s_white,v_white]
    if ((100<=h<=124) & (43<=s<=255) & (46<=v<=255)):
        img_hsv[:,:,0] = 0
        img_hsv[:,:,1] = 0
        img_hsv[:,:,2] = 255
    
    # Step2: for this region, we change their color 
    img_output = cv2.cvtColor(img_hsv,cv2.COLOR_HSV2RGB)
    return img_output
```

```python
img_change = backgroung_color_blue2white(img_blue)
plt.figure(figsize=(6, 6))
plt.imshow(img_change)
plt.show()
```

```python
def backgroung_color_blue2red(img):
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # for region that the value exceed a threshold, we specify it as the background color(blue)
    img_hsv[:,:,0] = 0
    img_output = cv2.cvtColor(img_hsv,cv2.COLOR_HSV2RGB)
    return img_output
```

```python
img_change = backgroung_color_blue2white(img_blue)
plt.figure(figsize=(6, 6))
plt.imshow(img_change)
plt.show()
```
